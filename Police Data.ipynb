```python
import pandas as pd

police=pd.read_csv(r"C:\Users\Harold\Desktop\python\projects\Dataset  Project 3  Police Data Analysis.csv")

print("Remove the missing values ")

#Show null values
null=police.isnull().sum()
print(null)

#Remove the null values
police.drop(columns='country_name',inplace=True)
print(police.columns)

print("For Speeeding, were men or Women stopped more often ?")
#Consider speeding and driver gender
#df[df.column1=='element/value'].column2.value_counts()
compare=police[police.violation == 'Speeding'].driver_gender.value_counts()
print(compare)

#Men were stopped more often at 25517 stops compared to the females' 11686

print("Does gender affects who gets searched during a stop ?")
#df.groupby('column1').column2.sum()#search_conducted/gender
search=police.groupby('driver_gender').search_conducted.sum()
print(search)
search1=print(f"The searches data was as follows \n{police.search_conducted.value_counts()}")
search
print("Of the 2479 searches conducted, 366 were for females and 2116 were on males implying gender is a factor influencing searches at police stops \n\n")

print("What is the mean stop duration? #Mapping and Datatype casting")
#df['column']=df['column'].map((old:new,old:new))#Mapping means we ma new values to the column value
#casting means changing the data type
#cnvert the str to int type
#df['column'].mean()
#Fill the null values using fillna()
police['stop_duration']=police['stop_duration'].fillna('')
#print(police['stop_duration'])
police['stop_duration']=police['stop_duration'].map({'0-15 Min':7.5,'16-30 Min':24,'30+ Min':45,'':0})
print(police['stop_duration'])
print(police['stop_duration'].value_counts())
#police['stop_duration'].mean()
print(f"The mean stop duration is {police['stop_duration'].mean()} minutes \n \n")

print("COMPARE THE AGE DISTRIBUTIONS FOR EACH VIOLATION")
#df.groupby('violation').age_distribution.describe()
ageDistribution=police.groupby('violation').driver_age.describe()
print(ageDistribution)

print("Data Summary of Age Violations")
print("Equipment violations were committed by suspects as young as 15 to as old as 81 \n")
print("Moving violations recorded at an average age of 47 as seen in the 50th percentile \n")
print("Speeding violations were recorded for drivers of the younger ages to highs of 88 year old drivers")

print(police.columns)


print("Compare driver race and mean stop duration for each race ")
print(police['driver_race'].value_counts())
meanStopbyRace=police.groupby('driver_race').stop_duration.describe()
print(meanStopbyRace)

print("From the data, it is evident that blacks and hispanics get stoped for longer, with stops lasting up to 23 minutes, \n as compared to their asian and white counterparts whose maximum stop times last no more than 8 minutes. \n Even though more whites get stopped by traffic police")
```
